# Homework: Convergence and Divergence of Series

In this assignment, we will explore two infinite series:
- A **convergent series**: $$\sum_{n=1}^{\infty} \frac{1}{n^2}.$$
- A **divergent series**: $$\sum_{n=1}^{\infty} \frac{1}{\sqrt{n}}.$$

We will first analyze these series symbolically and then estimate them by computing their partial sums.


## Series Analysis Using Ratio and Root Tests

We analyze the convergence of the following series using two common tests: the **ratio test** and the **root test**. In both cases, we compute the limits

1. **Ratio Test:**  
   $$
   L = \lim_{n \to \infty} \frac{a_{n+1}}{a_n}.
   $$

2. **Root Test:**  
   $$
   L' = \lim_{n \to \infty} \left(a_n\right)^{\frac{1}{n}}.
   $$

If either limit is less than 1, the series converges absolutely; if greater than 1, it diverges. If the limit equals 1, the test is inconclusive.

---

### Convergent Series:
$$
\sum_{n=1}^{\infty} \frac{1}{n^2}
$$

Let
$$
a_n = \frac{1}{n^2}.
$$

**Ratio Test:**

$$
\frac{a_{n+1}}{a_n} = \frac{\frac{1}{(n+1)^2}}{\frac{1}{n^2}} = \frac{n^2}{(n+1)^2}.
$$

Taking the limit:

$$
\lim_{n\to\infty} \frac{n^2}{(n+1)^2} = \lim_{n\to\infty} \frac{1}{\left(1 + \frac{1}{n}\right)^2} = 1.
$$

**Root Test:**

$$
\left(a_n\right)^{\frac{1}{n}} = \left(\frac{1}{n^2}\right)^{\frac{1}{n}} = \frac{1}{n^{\frac{2}{n}}} = e^{-\frac{2\ln n}{n}}
$$

To evaluate the limit

$$
\lim_{n\to\infty} \frac{2\ln n}{n},
$$

we apply L'Hôpital's rule. Since

$$
\lim_{n\to\infty} \frac{\ln n}{n} = \lim_{n\to\infty} \frac{1/n}{1} = 0,
$$

it follows that

$$
\lim_{n\to\infty} e^{-\frac{2\ln n}{n}} = e^{0} = 1.
$$

**Conclusion for**
$$
\sum_{n=1}^{\infty} \frac{1}{n^2}:
$$  
Both tests yield 1 (inconclusive). However, since this is a **p-series** with $p = 2 > 1 $, we know it converges.

---

### Divergent Series:
$$
\sum_{n=1}^{\infty} \frac{1}{\sqrt{n}} = \sum_{n=1}^{\infty} \frac{1}{n^{1/2}}
$$

Let
$$
a_n = \frac{1}{\sqrt{n}}.
$$

**Ratio Test:**

$$
\frac{a_{n+1}}{a_n} = \frac{\frac{1}{\sqrt{n+1}}}{\frac{1}{\sqrt{n}}} = \sqrt{\frac{n}{n+1}}.
$$

Taking the limit:

$$
\lim_{n\to\infty} \sqrt{\frac{n}{n+1}} = 1.
$$

**Root Test:**

$$
\left(a_n\right)^{\frac{1}{n}} = \left(\frac{1}{n^2}\right)^{\frac{1}{n}} = \frac{1}{n^{\frac{2}{n}}} = e^{-\frac{2\ln n}{n}}
$$

Again, since

$$
\lim_{n\to\infty} \frac{\ln n}{2n} = 0,
$$

we obtain

$$
\lim_{n\to\infty} e^{-\frac{\ln n}{2n}} = e^{0} = 1.
$$

**Conclusion for**
$$
\sum_{n=1}^{\infty} \frac{1}{\sqrt{n}}:
$$  
Both tests yield 1 (inconclusive)


## P-Series Test

A **p-series** is a series of the form
$$
\sum_{n=1}^{\infty} \frac{1}{n^p}.
$$

The p-series test states that:

- The series **converges** if \($p > 1$).
- The series **diverges** if \($p \leq 1$).

Note that this is a direct result of the **integral-test** that you saw in the lecture.

This test provides a quick and effective way to determine the convergence of series whose terms decrease as a power of \(n\).


## Estimating the Series via Partial Sums

We can approximate an infinite series by computing its **partial sum**:
$$
S(g) = \sum_{n=1}^{g} a_n,
$$
where $g$ is the number of terms used in the approximation.

- For a convergent series, as $ g \to \infty $, the partial sum $S(g)$ approaches the true value of the series.
- For a divergent series, the partial sum $S(g)$ will grow without bound (or fail to settle) as $g \to \infty$.

In the next cell, we will compute these partial sums using Python and observe how the sums behave as $g$ increases.


```python
# Python code to compute and plot the partial sums of the series

import numpy as np
import matplotlib.pyplot as plt

def partial_sum(series_func, N):
    """Compute the partial sum of a series for N terms."""
    n = np.arange(1, N+1)
    return np.sum(series_func(n))

# Define the series functions
def convergent_series(n):
    return 1 / n**2

def divergent_series(n):
    return 1 / np.sqrt(n)

# List of g values (number of terms)
terms = [10, 100, 1000, 10000, 100000]

# Compute partial sums for both series
convergent_sums = [partial_sum(convergent_series, N) for N in terms]
divergent_sums = [partial_sum(divergent_series, N) for N in terms]

# Display the results
print("Partial sums for the convergent series ∑ 1/n^2:")
for N, s in zip(terms, convergent_sums):
    print(f"g = {N:6d} : S(g) = {s:.6f}")

print("\nPartial sums for the divergent series ∑ 1/√n:")
for N, s in zip(terms, divergent_sums):
    print(f"g = {N:6d} : S(g) = {s:.6f}")

# Plotting the partial sums
plt.figure(figsize=(12, 5))

# Convergent series plot
plt.subplot(1, 2, 1)
plt.plot(terms, convergent_sums, marker='o')
plt.xscale('log')
plt.xlabel('Number of Terms (g)')
plt.ylabel('Partial Sum S(g)')
plt.title('Convergent Series: ∑ 1/n^2')

# Divergent series plot
plt.subplot(1, 2, 2)
plt.plot(terms, divergent_sums, marker='o', color='red')
plt.xscale('log')
plt.xlabel('Number of Terms (g)')
plt.ylabel('Partial Sum S(g)')
plt.title('Divergent Series: ∑ 1/√n')

plt.tight_layout()
plt.show()
```

## Example 1: Geometric Series

Consider the power series:
$$
f(x) = \sum_{n=0}^{\infty} x^n,
$$
with coefficients $a_n = 1$ for all $n$.

**Using the Ratio Test:**

We compute
$$
L = \lim_{n\to\infty} \left|\frac{a_{n+1} \, x^{n+1}}{a_n \, x^n}\right|
= \lim_{n\to\infty} \left|\frac{1 \cdot x^{n+1}}{1 \cdot x^n}\right|
= \lim_{n\to\infty} |x|
= |x|.
$$

For the series to converge, we require $L < 1$ , i.e.,
$$
|x| < 1.
$$

Thus, the radius of convergence is
$$
R = 1.
$$


## Example 2: Exponential Series

Consider the power series:
$$
f(x) = \sum_{n=0}^{\infty} \frac{x^n}{n!},
$$
with coefficients $a_n = \frac{1}{n!}$.

**Using the Ratio Test:**

We compute
$$
L = \lim_{n\to\infty} \left|\frac{a_{n+1} \, x^{n+1}}{a_n \, x^n}\right|
= \lim_{n\to\infty} \left|\frac{\frac{1}{(n+1)!} \, x^{n+1}}{\frac{1}{n!} \, x^n}\right|
= \lim_{n\to\infty} \left|\frac{x}{n+1}\right|
= 0.
$$

Since $L = 0 < 1$ for all $x$, the series converges for every real number. Therefore, the radius of convergence is
$$
R = \infty.
$$

## Taylor Series Expansion of $\ln(x) $ about $x_0 = 1 $

We derive the Taylor series for
$$
f(x) = \ln(x)
$$

around $ x_0 = 1 $ . Recall that the Taylor series for a function $ f(x) $ about $ x_0 $ is given by:

$$
f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(x_0)}{n!}\,(x - x_0)^n.
$$

### Step 1: Compute the Derivatives of $\ln(x)$
- $f(x) = \ln(x)$
- $f'(x) = \dfrac{1}{x}$
- $f''(x) = -\dfrac{1}{x^2}$
- $f'''(x) = \dfrac{2}{x^3}$
- $f^{(4)}(x) = -\dfrac{6}{x^4}$

In general, for $n \ge 1$, the $n$th derivative of $\ln(x)$ is:
$$
f^{(n)}(x) = (-1)^{n-1} (n-1)! \cdot x^{-n}.
$$

### Step 2: Evaluate the Derivatives at $x_0 = 1$
- $f(1) = \ln(1) = 0$
- $f'(1) = \dfrac{1}{1} = 1$
- $f''(1) = -\dfrac{1}{1^2} = -1$
- $f'''(1) = \dfrac{2}{1^3} = 2$
- $f^{(4)}(1) = -\dfrac{6}{1^4} = -6$


Thus, for $ n \ge 1 $:
$$
f^{(n)}(1) = (-1)^{n-1} (n-1)!.
$$

### Step 3: Formulate the Taylor Series about $ x_0 = 1 $
Substitute the values into the Taylor series formula:
$$
\ln(x) = f(1) + f'(1)(x-1) + \frac{f''(1)}{2!}(x-1)^2 + \frac{f'''(1)}{3!}(x-1)^3 + \frac{f^{(4)}(1)}{4!}(x-1)^4 + \cdots.
$$

Since $ f(1) = 0 $, we have:
$$
\ln(x) = 1\cdot (x-1) + \frac{-1}{2}(x-1)^2 + \frac{2}{6}(x-1)^3 + \frac{-6}{24}(x-1)^4 + \cdots.
$$

Simplify the coefficients:
$$
-  \frac{2}{6} = \frac{1}{3}
-  \frac{-6}{24} = -\frac{1}{4}
$$
So the series becomes:

$$
\ln(x) = (x-1) - \frac{1}{2}(x-1)^2 + \frac{1}{3}(x-1)^3 - \frac{1}{4}(x-1)^4 + \cdots.
$$

### Convergence
This Taylor series converges for:
$$
|x-1| < 1.
$$

Thus, the Taylor series expansion of $ \ln(x) $ about $ x_0 = 1 $ is:
$$
\boxed{\ln(x) = (x-1) - \frac{1}{2}(x-1)^2 + \frac{1}{3}(x-1)^3 - \frac{1}{4}(x-1)^4 + \cdots, \quad \text{for } |x-1| < 1.}
$$

A p-series has the form

### **1B. Integral Test**

Consider the function
$$
f(x) = \frac{1}{x},
$$
which is positive and decreasing for \(x \ge 1\). The corresponding improper integral is
$$
\int_{1}^{\infty} \frac{1}{x} \, dx.
$$
1. **Compute the antiderivative**:
   $$
   \int \frac{1}{x}\,dx = \ln|x|.
   $$
2. **Evaluate the improper integral**:
   $$
   \lim_{M \to \infty}
   \int_{1}^{M} \frac{1}{x}\,dx
   =
   \lim_{M \to \infty} \bigl[\ln(x)\bigr]_1^M
   =
   \lim_{M \to \infty} \left( \ln(M) - \ln(1) \right)
   =
   \lim_{M \to \infty} \ln(M)
   =
   \infty.
   $$

Since the integral diverges, the **Integral Test** implies the series also diverges.

**Conclusion for Series 1**:$\sum_{n=1}^{\infty}\frac{1}{n}$ **diverges**.

## McLaurin Series Definition and Derivation

The McLaurin series for a function $f(x)$ is the Taylor series expansion of $f(x)$ about $x=0$. It is given by:
$$
f(x)=\sum_{n=0}^{\infty} \frac{f^{(n)}(0)}{n!}\, x^n,
$$
where $f^{(n)}(0)$ denotes the $n$th derivative of $f(x)$ evaluated at $x=0$.

---

### Example 1: Deriving the McLaurin Series for $e^x$

Let $f(x) = e^x$.

**1. Find the derivatives:**

\[
\begin{aligned}
f(x) &= e^x, \\
f'(x) &= e^x, \\
f''(x) &= e^x, \\
&\ \vdots \\
\text{In general, for all } n:&\quad f^{(n)}(x)= e^x.
\end{aligned}
\]

**2. Evaluate at $x=0$:**

$$
f^{(n)}(0) = e^0 = 1.
$$

**3. Substitute into the McLaurin series formula:**

$$
e^x = \sum_{n=0}^{\infty} \frac{1}{n!}\, x^n.
$$

**4. Determine the radius of convergence:**

Using the ratio test:
$$
L = \lim_{n\to\infty} \left|\frac{x^{n+1}/(n+1)!}{x^n/n!}\right|
= \lim_{n\to\infty} \left|\frac{x}{n+1}\right|
= 0,
$$
for any finite $x$. Thus, the series converges for all $x$, and the radius of convergence is:
$$
R = \infty.
$$

---

### Example 2: Deriving the McLaurin Series for $\ln(1+x)$

Let $f(x) = \ln(1+x)$.

**1. Find the derivatives:**

$$ f(x) = \ln(1+x) $$

$$ f(0) = \ln(1) = 0 $$

$$ f'(x) = \frac{1}{1+x} $$

$$ f'(0) = 1 $$

$$ f''(x) = -\frac{1}{(1+x)^2} $$

$$ f''(0) = -1 $$

$$ f'''(x) = \frac{2}{(1+x)^3} $$

$$ f'''(0) = 2 $$

$$ f^{(4)}(x) = -\frac{6}{(1+x)^4} $$

$$ f^{(4)}(0) = -6 $$


**2. Identify the pattern:**

For $n \geq 1$, the $n$th derivative at 0 is:
$$
f^{(n)}(0) = (-1)^{n-1} (n-1)!.
$$

**3. Substitute into the McLaurin series formula:**

$$
\ln(1+x) = \sum_{n=1}^{\infty} \frac{f^{(n)}(0)}{n!}\, x^n
= \sum_{n=1}^{\infty} \frac{(-1)^{n-1}(n-1)!}{n!}\, x^n.
$$

Since
$$
\frac{(n-1)!}{n!} = \frac{1}{n},
$$
this simplifies to:
$$
\ln(1+x) = \sum_{n=1}^{\infty} (-1)^{n-1} \frac{x^n}{n}.
$$

**4. Determine the radius of convergence:**

By applying the ratio test, it can be shown that the series converges for $|x| < 1$. Therefore, the radius of convergence is:
$$
R = 1.
$$

These derivations illustrate the process of obtaining the McLaurin series by computing the derivatives at $x=0$ and substituting them into the Taylor series formula.

## Problem 2: Taylor Series for $e^x$ about $x = 1$

We want to expand $e^x$ in a Taylor series around $x = 1$. By definition, the Taylor series of a function $f(x)$ about $x = x_0$ is:

$$
f(x)
= \sum_{n=0}^{\infty}
\frac{f^{(n)}(x_0)}{n!} (x - x_0)^n.
$$

Here, $f(x) = e^x$ and $x_0 = 1$.

### Step 1: Compute the derivatives and evaluate at $x_0=1$

- $f(x) = e^x$.
- $f'(x) = e^x$.
- $f''(x) = e^x$.  
- All higher derivatives are also $e^x$.

At $x = 1$:

- $f(1) = e^1 = e$.
- $f'(1) = e^1 = e$.
- $f''(1) = e^1 = e$.
- In general, $f^{(n)}(1) = e$ for every $n$.

### Step 2: Plug into the Taylor series formula

Thus,

$$
e^x
= \sum_{n=0}^{\infty}
\frac{e}{n!} (x - 1)^n.
$$

### Step 3: Write out the first three nonzero terms

The term for $n=0$ is
$$
\frac{e}{0!}(x - 1)^0 = e.
$$

For $n=1$:
$$
\frac{e}{1!}(x - 1)^1 = e (x - 1).
$$

For $n=2$:
$$
\frac{e}{2!}(x - 1)^2 = \frac{e}{2} (x - 1)^2.
$$

So, up to the first three nonzero terms, the Taylor expansion about $x=1$ is:
$$
e^x \approx e + e (x - 1) + \frac{e}{2} (x - 1)^2 + \cdots
$$

# Howe accurate these power series estimations are? Lets compare them in the followin plots

```python
import numpy as np
import matplotlib.pyplot as plt

# Define functions

# For e^x: Exact and Taylor expansion about x=1 using 3 nonzero terms
def exact_exp(x):
    return np.exp(x)

def taylor_exp(x):
    # Taylor expansion about x=1: e + e*(x-1) + (e/2)*(x-1)^2
    return np.e + np.e*(x - 1) + (np.e/2)*(x - 1)**2

# For sin(x): Exact and McLaurin series using 3 nonzero terms
def exact_sin(x):
    return np.sin(x)

def mclaurin_sin(x):
    # McLaurin series for sin(x): x - x^3/6 + x^5/120
    return x - (x**3)/6 + (x**5)/120

# For cos(x): Exact and McLaurin series using 3 nonzero terms
def exact_cos(x):
    return np.cos(x)

def mclaurin_cos(x):
    # McLaurin series for cos(x): 1 - x^2/2 + x^4/24
    return 1 - (x**2)/2 + (x**4)/24

# Define x ranges for plotting
x_exp = np.linspace(0.5, 1.5, 400)  # for e^x expansion around x=1
x_trig = np.linspace(-2, 2, 400)      # for sin(x) and cos(x) approximations near 0

# Create subplots
plt.figure(figsize=(12, 12))

# Plot for e^x
plt.subplot(3, 1, 1)
plt.plot(x_exp, exact_exp(x_exp), label='Exact $e^x$', color='blue')
plt.plot(x_exp, taylor_exp(x_exp), label='Taylor Approx. (3 terms)', linestyle='--', color='red')
plt.title('Taylor Series Approximation for $e^x$ about $x=1$')
plt.xlabel('x')
plt.ylabel('$e^x$')
plt.legend()
plt.grid(True)

# Plot for sin(x)
plt.subplot(3, 1, 2)
plt.plot(x_trig, exact_sin(x_trig), label='Exact $\sin(x)$', color='blue')
plt.plot(x_trig, mclaurin_sin(x_trig), label='McLaurin Approx. (3 terms)', linestyle='--', color='red')
plt.title('McLaurin Series Approximation for $\sin(x)$')
plt.xlabel('x')
plt.ylabel('$\sin(x)$')
plt.legend()
plt.grid(True)

# Plot for cos(x)
plt.subplot(3, 1, 3)
plt.plot(x_trig, exact_cos(x_trig), label='Exact $\cos(x)$', color='blue')
plt.plot(x_trig, mclaurin_cos(x_trig), label='McLaurin Approx. (3 terms)', linestyle='--', color='red')
plt.title('McLaurin Series Approximation for $\cos(x)$')
plt.xlabel('x')
plt.ylabel('$\cos(x)$')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
