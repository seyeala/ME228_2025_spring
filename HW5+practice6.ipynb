# Section 0: ODEs in Physics and Engineering

Ordinary differential equations (ODEs) are a powerful tool for modeling how physical
quantities. Many natural laws and principles
in physics can be expressed using ODEs. For example, Newton’s second law
\\( F = m \, a \\) leads to the equation of motion for objects by relating forces
to acceleration. By formulating such laws as differential equations, we can
**simulate** the dynamic behavior of systems – predict how they evolve, respond
to inputs, or come into equilibrium.

ODE-based models can be used to design systems and understand
phenomena such the vibrations, heat transfer, and mechanics of fluids .
A classic example is a **mass oscillation** on a spring, which are covered in your textbook ( chapter 2.4 and 2.8). Note that your dynamics, and vibration courses you will study this phenomenon throughly. However, here we mention it briefly as an example of the utility of ODEs.

If a mass \\(m\\) is attached
to an ideal spring with the stiffness \\(k\\) and displaced from equilibrium, Hooke’s law
requires a restoring spring force \\(-k\,x\\) that moves the mass back toward
equilibrium. Neglecting damping the motion is governed by this ODE:

**Equation (1):**
$$
m \,\frac{d^2 x}{dt^2} + k\,x = 0
$$

Equation (1) is a simple harmonic oscillator. Its solution is oscillatory
motion (sine/cosine) at the natural frequency \\( \omega_n = \sqrt{k/m} \\). This ODE model is an example of countless vibrational systems like vibration of molecules or oscilation of a pandulum.

Another example is **heat transfer** in a one-dimensional rod. The temperature
\\( T(x,t) \\) in a rod evolving over time \\( t \\) and along position \\( x \\) can be
governed by the heat equation:

**Equation (2):**
$$
\frac{\partial T}{\partial t} \;=\; \alpha \,\frac{\partial ^2 T}{\partial x^2}
$$

In summary, ODEs form the backbone of modeling
in mechanics, thermodynamics, circuits, and many other fields, allowing us to predict and design the behavior of real-world systems.

# Section 1 : Deriving and Solving the Governing Equation

We now focus on a **mass-spring-damper** system. Let a mass \\(m\\) be attached to a spring (stiffness \\(k\\)) and a damper (coefficient \\(c\\)). An external force \\(f(t)\\) acts on the mass. The derivation below follows Newton’s second law with a force balance.

---

# Section 2 <font color="red">Practice 6.</font>: Newton’s Second Law and Acceleration

Newton’s second law states that the net force acting on an object is equal to its mass times its acceleration.

**Equation (1): Newton’s Second Law**

$$
F_{\text{total}} = m\,a.
\tag{1}
$$

The acceleration \\(a\\) by definition is the second derivative of the displacement \\(x(t)\\) with respect to time ( by definition velocity is the first and accerleration is the second dervitive of displacment):

**Equation (2): Definition of Acceleration**

$$
a = \frac{d^\boxed{\text{Q1-1}} x}{dt^\boxed{\text{Q1-1}}}.
\tag{2}
$$

Substituting Equation (2) into Equation (1):

**Equation (1a): Newton’s Law in Terms of \\(x(t)\\)**

$$
F_{\text{total}} = m\,\frac{d^2 x}{dt^2}.
\tag{1a}
$$

---

#Section 3 <font color="red">Practice 6.</font>: Establishing the Net Force

Assume that:
- \\(x(t)\\) is the displacement of the mass.
- The external force \\(f(t)\\) acts to the right (positive direction).

The mass experiences opposing forces:
- **Spring force:** A restoring force given by Hooke’s law.
- **Damping force:** A force proportional to the velocity from the damper.

Thus, the net force is given by:

**Equation (3): Net Force on the Mass**

$$
F_{\text{total}} = f(t) - F_{\text{spring}} - F_{\text{damper}}.
\tag{3}
$$

The spring force is defined by:

**Equation (4): Spring Force**

$$
F_{\text{spring}} = k\,x.
\tag{4}
$$

And the damping force is:

**Equation (5): Damping Force**

$$
F_{\text{damper}} = c\,\frac{dx}{dt}.
\tag{5}
$$

---

#Section 4 <font color="red">Practice 6.</font>: Forming the Differential Equation

Substitute Equations (4) and (5) into Equation (3):

$$
F_{\text{total}} = f(t) - \boxed{\text{Q1-2}}\,x - c\,\frac{dx}{dt}.
$$

Then, equate this expression with Newton’s law (Equation (1a)):

**Equation (6): Governing Equation Before Rearrangement**

$$
m\,\frac{d^2 x}{dt^2} = f(t) - c\,\frac{dx}{dt} - k\,x.
\tag{6}
$$

Rearrange Equation (6) to obtain the standard form of a second order ODE that you are familiar with:

**Equation (7): Governing Differential Equation**

$$
m\,\frac{d^2 x}{dt^2} + c\,\frac{dx}{dt} + k\,x = f(t).
\tag{7}
$$

---

#Section 5 <font color="red">Practice 6.</font>: Solving the Homogeneous Equation (AKA Free Oscilation) (\\(f(t)=0\\))

Consider the homogeneous case by setting \\(f(t)=0\\):

**Equation (8): Homogeneous Differential Equation**

$$
m\,\frac{d^2 x}{dt^2} + \boxed{\text{Q1-3}}\,\frac{dx}{dt} + k\,x = 0.
\tag{8}
$$

Assume a trial solution of the form:

$$
x_h(t) = e^{rt}.
$$

Then, its derivatives are:
- First derivative:
  $$
  \frac{dx_h}{dt} = r\,e^{rt},
  $$
- Second derivative:
  $$
  \frac{d^2x_h}{dt^2} = r^2\,e^{rt}.
  $$

Substitute these into Equation (8):

$$
m\,\boxed{\text{Q1-4}}^2\,e^{rt} + c\,r\,e^{rt} + k\,e^{rt} = 0.
$$

Since \\(e^{rt} \neq 0\\), divide the entire equation by \\(e^{rt}\\):

**Equation (9): Characteristic Equation**

$$
m\,r^2 + \boxed{\text{Q1-5}}\,r + k = 0.
\tag{9}
$$

---

# Section 6: Analyzing the Characteristic Equation

The discriminant for the characteristic equation is:

$$
\Delta = c^\boxed{\text{Q1-6}} - 4\boxed{\text{Q1-7}}k.
$$

We now consider three cases based on the value of \\(\Delta\\) as we learned to do with the 2nd order ODEs with constant coefficents:

1. **Case I (In the context of dynamics it has a specific name Overdamped):** If \\(\Delta > \boxed{\text{Q1-8}}\\), the quadratic equation has two distinct real roots:

   **Equation (10): Overdamped Roots**

   $$
   r_{1,2} = \frac{-\boxed{\text{Q1-9}} \pm \sqrt{c^2 - 4mk}}{2m}.
   \tag{10}
   $$

2. **Case II (In the context of dynamics it has a specific name Critically Damped):** If \\(\Delta = 0\\), the quadratic equation has a repeated real root:

   **Equation (11): Critically Damped Root**

   $$
   r = -\frac{c}{2\boxed{\text{Q1-10}}}.
   \tag{11}
   $$

3. **Case III (In the context of dynamics it has a specific name Underdamped):** If \\(\Delta < 0\\), the quadratic equation has two complex conjugate roots. Express these as:

   **Equation (12): Underdamped Roots**

   $$
   r_{1,2} = -\frac{c}{2m} \pm i\,\omega_d, \quad \text{with} \quad \omega_d = \sqrt{\frac{k}{\boxed{\text{Q1-11}}} - \left(\frac{c}{2m}\right)^2}.
   \tag{12}
   $$

You may have noticed that you learned to solve ordinary differential equations (ODEs) by classifying them according to their discriminant, regardless of their application. However, when dealing with mass-spring systems, these classifications acquire specific names—overdamped, underdamped, and critically damped. I would like to emphasize that the abstraction of mathematics is one of its greatest strengths. # Section 7: Homogeneous Solution for Different Damping Cases

Based on the nature of the roots founf in Section 5, the homogeneous solution \\(x_h(t)\\) takes different forms:

1. **Case I (Overdamped, \\(\Delta > 0\\))**

   With distinct real roots \\(r_1\\) and \\(r_2\\) from Equation (10):

   **Equation (13): Overdamped Homogeneous Solution**

   $$
   x_h(t) = A\,e^{r_1t} + B\,e^{r_2t},
   \tag{13}
   $$

   where \\(A\\) and \\(B\\) are constants determined by initial conditions.

2. **Case II (Critically Damped, \\(\Delta = 0\\))**

   With a repeated real root \\(r = -\frac{c}{2m}\\) from Equation (11):

   **Equation (14): Critically Damped Homogeneous Solution**

   $$
   x_h(t) = \left(A + Bt\right)\boxed{\text{Q1-12}}^{-\frac{c}{2m}t},
   \tag{14}
   $$

   where \\(A\\) and \\(B\\) are constants determined by initial conditions.

3. **Case III (Underdamped, \\(\Delta < 0\\))**

   With complex conjugate roots from Equation (12):

   **Equation (15): Underdamped Homogeneous Solution**

   $$
   x_h(t) = e^{-\frac{c}{2m}t}\left[A\,\cos(\omega_d t) + B\,\sin(\omega_d t)\right],
   \tag{15}
   $$

   where \\(A\\) and \\(B\\) are constants determined by initial conditions.

---

#Section 8: Finding the Particular Solution

For a constant external force \\(f(t) = F_0\\), we look for a constant particular solution \\(x_p\\). Since the derivatives of a constant are zero:

$$
\frac{dx_p}{dt} = 0 \quad \text{and} \quad \frac{d^2x_p}{dt^2} = 0,
$$

substitute into Equation (7):

**Equation (16): Particular Solution Condition**

$$
m\,(0) + c\,(0) + k\,x_p = F_0.
\tag{16}
$$

This simplifies to:

**Equation (17): Particular Solution**

$$
x_p = \frac{F_0}{k}.
\tag{17}
$$

---

#Section 9: Combining the Solutions AKA (Forced Vibration)

The general solution of the nonhomogeneous differential equation (Equation (7)) is the sum of the homogeneous solution \\(x_h(t)\\) and the particular solution \\(x_p\\):

**Equation (18): General Solution**

$$
x(t) = x_h(t) + x_p.
\tag{18}
$$

For example, in the underdamped case (Case III), the general solution becomes:

$$
x(t) = e^{-\frac{c}{2m}t}\left[A\,\cos(\omega_d t) + B\,\sin(\omega_d t)\right] + \frac{F_0}{k}.
$$

The constants \\(A\\) and \\(B\\) are determined by the initial conditions \\(x(0)=x_0\\) and \\(x'(0)=v_0\\).

A typical final form for the underdamped, constant-force case is expressed as:

**Equation (19): Final Form of the Solution (Underdamped Example)**

$$
x(t) = \frac{F_0}{k} + e^{-\frac{c}{2m}\boxed{\text{Q1-13}}}\left[\left(x_0 - \frac{F_0}{k}\right)\cos(\omega_d \boxed{\text{Q1-14}}) + \frac{v_0 + \frac{c}{2m}\left(x_0 - \frac{F_0}{k}\right)}{\omega_d}\sin(\omega_d t)\right].
\tag{19}
$$

In the next section, we choose specific parameter values and plot \\(x(t)\\) over time.


# Section 10: Step-by-Step Solution for Specific Parameters

Lets now solve the governing differential equation (Equation (7) from the previous section) with the following parameters:

- $ m = 1.0 $ kg  
- $ k = 20.0 $ N/m  
- $ c = 2.0 $ N·s/m  
- $ F_0 = 10.0 $ N  
- $ x(0) = x_0 = 0.0 $ m  
- $ \dot{x}(0) = v_0 = 0.0 $ m/s  

From the derivation above, the particular solution for a constant force (see Equation (17)) is

**Equation (20): Particular Solution**

$$
x_p = \frac{F_0}{k} = \frac{10}{20} = 0.5 \,\text{m}.
\tag{20}
$$

Hence, the general solution for the forced system (see Equation (14)) is written as

**Equation (21): General Solution**

$$
x(t) = x_p + e^{-\frac{c}{2m}t}\left[A\,\cos(\omega_d t) + B\,\sin(\omega_d t)\right].
\tag{21}
$$

---

### 1. Calculate $\omega_d$

First, the natural frequency is given by

**Equation (22): Natural Frequency**

$$
\omega_n = \sqrt{\frac{k}{m}} = \sqrt{\frac{20}{1}} = \sqrt{20} \approx 4.4721.
\tag{22}
$$

The damped natural frequency is then

**Equation (23): Damped Natural Frequency**

$$
\omega_d = \sqrt{\omega_n^2 - \left(\frac{c}{2m}\right)^2}
= \sqrt{20 - \left(\frac{2}{2\cdot1}\right)^2}
= \sqrt{20 - 1}
= \sqrt{19} \approx 4.3588.
\tag{23}
$$

---

### 2. Find \(A\) and \(B\)

Using the initial condition for displacement at \(t=0\):

**Equation (24): Initial Displacement Condition**

$$
x(0) = x_p + A = 0.5 + A.
\tag{24}
$$

Since \( x(0) = 0.0 \), we have

**Equation (25): Determination of \(A\)**

$$
0.5 + A = 0 \quad \Longrightarrow \quad A = -0.5.
\tag{25}
$$

Next, differentiate Equation (21) to obtain the velocity $\dot{x}(t)$. Let
$
\alpha = \frac{c}{2m} = \frac{2}{2\cdot1} = 1.
$
Differentiating, we have

**Equation (26): Derivative of the General Solution**

$$
\dot{x}(t) = \frac{d}{dt}\left[x_p + e^{-\alpha t}\Bigl(A\,\cos(\omega_d t) + B\,\sin(\omega_d t)\Bigr)\right].
\tag{26}
$$

Since $x_p$ is constant, its derivative is zero. Applying the product rule to the second term gives

**Equation (27): Expanded Derivative**

$$
\dot{x}(t) = -\alpha\,e^{-\alpha t}\Bigl[A\,\cos(\omega_d t) + B\,\sin(\omega_d t)\Bigr] + e^{-\alpha t}\Bigl[-A\,\omega_d\,\sin(\omega_d t) + B\,\omega_d\,\cos(\omega_d t)\Bigr].
\tag{27}
$$

At $t = 0$, note that $ e^{-\alpha t} = 1 $, $\cos(0) = 1$, and $\sin(0) = 0$, so

**Equation (28): Initial Velocity Expression**

$$
\dot{x}(0) = -\alpha A + B\,\omega_d.
\tag{28}
$$

Given that $\dot{x}(0) = 0$ and $\alpha = 1$:

**Equation (29): Velocity Condition at \(t=0\)**

$$
0 = -1\cdot(-0.5) + B\,\omega_d = 0.5 + B\,\omega_d.
\tag{29}
$$

Thus, we solve for \(B\):

**Equation (30): Determination of \(B\)**

$$
B\,\omega_d = -0.5 \quad \Longrightarrow \quad B = -\frac{0.5}{\omega_d}.
\tag{30}
$$

Substituting $\omega_d \approx 4.3588$:

**Equation (31): Numerical Value for \(B\)**

$$
B \approx -\frac{0.5}{4.3588} \approx -0.1147.
\tag{31}
$$

---

### 3. Final Solution

Substitute the values $ A = -0.5 $ and $ B \approx -0.1147 $ back into Equation (21):

**Equation (32): Final Solution (Numerical Form)**

$$
x(t) = 0.5 + e^{-t}\Bigl[-0.5\,\cos(4.3588\,t) - 0.1147\,\sin(4.3588\,t)\Bigr].
\tag{32}
$$

In exact terms, using $\omega_d = \sqrt{19}$, the final solution can be written as:

**Equation (33): Final Solution (Exact Form)**

$$
x(t) = 0.5 + e^{-t}\left[-0.5\,\cos\left(\sqrt{19}\,t\right) - \frac{0.5}{\sqrt{19}}\,\sin\left(\sqrt{19}\,t\right)\right].
\tag{33}
$$

Therefore, the mass starts at zero displacement with zero velocity and then oscillates around \(0.5\) m with an exponentially decaying amplitude.
Lets plot this solution next.

```python
import numpy as np
import matplotlib.pyplot as plt

# Given physical parameters and initial conditions
m = 1.0      # mass (kg)
k = 20.0     # spring constant (N/m)
c = 2.0      # damping coefficient (N·s/m)
F0 = 50.0    # external force magnitude (N)

x0 = 0.0     # initial displacement (m)
v0 = 0.0     # initial velocity (m/s)

# Derived constants
omega_n = np.sqrt(k/m)                        # natural frequency (rad/s)
omega_d = np.sqrt(k/m - (c/(2*m))**2)         # damped frequency (rad/s) for underdamped case

# Time range for simulation (0 to 8 seconds)
t = np.linspace(0, 8, 801)  # 801 points -> 0.01 s time step

# Compute the analytical solution x(t) using Equation (8)
x_p = F0 / k  # particular solution (static equilibrium)
# Homogeneous solution part:
A = x0 - x_p
B = (v0 + (c/(2*m)) * (x0 - x_p)) / omega_d
# Complete solution:
x_t = x_p + np.exp(-c/(2*m) * t) * (A * np.cos(omega_d * t) + B * np.sin(omega_d * t))

# Plot the displacement over time
plt.figure(figsize=(6,4))
plt.plot(t, x_t, label='x(t) displacement', color='orange')
plt.axhline(y=x_p, color='red', linestyle='--', label=f'Equilibrium = {x_p:.2f} m')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.title('Mass-Spring-Damper Response')
plt.grid(True)
plt.legend()
plt.show()
```

Below, in Section 10, you can enter your own values for mass m, damping c, spring
stiffness k, external force $F_0$, and initial conditions $x_0$ and $v_0$. Then,
the code will recalculate the solution and plot it, allowing you to explore different
parameter regimes. Note that they are based on Equation 19 which does not cover all possiblities (based on the discriminant)


```python
# Section 6: User Input for parameters and re-plot

import numpy as np
import matplotlib.pyplot as plt

def solve_mass_spring_damper(m, c, k, F0, x0, v0, t_end=8.0):
    """
    Solves the mass-spring-damper ODE:
      m x''(t) + c x'(t) + k x(t) = F0
    with initial conditions x(0)=x0, x'(0)=v0,
    and plots the displacement over time.
    """
    # Derived frequencies
    omega_n = np.sqrt(k / m)
    disc = omega_n**2 - (c/(2*m))**2
    if disc < 0.0:
        print("Warning: System is overdamped or critically damped, no sinusoidal part.")
        omega_d = np.sqrt(abs(disc)) # purely imaginary case
    else:
        omega_d = np.sqrt(disc)

    # Particular solution offset
    x_p = F0 / k  # shift

    # Solve for A, B via initial conditions
    # A = x0 - x_p
    A = x0 - x_p
    # 0 = v0 = x'(0) = - (c/(2m))*A + B*omega_d => B = ((c/(2m))*A + v0)/omega_d
    # but v0=0 in many cases; let's keep it general.
    if omega_d != 0.0:
        B = (v0 + (c/(2*m))*A) / omega_d
    else:
        # Avoid division by zero if disc=0 => critical damping
        B = 0.0

    # Time domain
    t_pts = 801
    t = np.linspace(0, t_end, t_pts)

    # Exponential term
    alpha = c/(2*m)
    # x(t) = x_p + e^{-alpha t}[ A cos(omega_d t) + B sin(omega_d t) ]
    if disc > 0.0:
        # underdamped or slightly underdamped
        x_t = x_p + np.exp(-alpha * t) * (
            A * np.cos(omega_d * t) + B * np.sin(omega_d * t)
        )
    elif disc == 0.0:
        # critically damped => x(t) = x_p + e^{-alpha t} [ (A + B t) ]
        # B can be found by matching x'(0)=v0
        # For simplicity let's do approximate
        x_t = x_p + np.exp(-alpha * t) * (A + B * t)
    else:
        # overdamped => let lamb1, lamb2 be real roots => do a simpler approach
        # For demonstration, let's do a quick numerical approach:
        lamb1 = alpha + np.sqrt(alpha**2 - k/m)
        lamb2 = alpha - np.sqrt(alpha**2 - k/m)
        # general solution = x_p + C1 e^{-lamb1 t} + C2 e^{-lamb2 t}
        # use x(0)=x0, x'(0)=v0 to solve for C1, C2
        # We'll do a short solve:
        # x(0)= x_p + C1 + C2 = x0
        # x'(t)=  -lamb1 C1 e^{-lamb1 t} - lamb2 C2 e^{-lamb2 t}
        # so x'(0)= -lamb1 C1 - lamb2 C2 = v0
        # Solve for C1, C2
        C1C2 = np.linalg.solve(
            np.array([[1, 1],[-lamb1, -lamb2]]),
            np.array([x0 - x_p, v0])
        )
        C1, C2 = C1C2[0], C1C2[1]
        x_t = x_p + C1*np.exp(-lamb1*t) + C2*np.exp(-lamb2*t)

    # Plot
    plt.figure(figsize=(6,4))
    plt.plot(t, x_t, label='x(t)', color='blue')
    plt.axhline(y=x_p, color='red', linestyle='--',
                label=f'Equilibrium = {x_p:.2f} m')
    plt.title('Mass-Spring-Damper Response (User Input)')
    plt.xlabel('Time (s)')
    plt.ylabel('Displacement (m)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Prompt user for inputs:
print("Enter the mass-spring-damper parameters:")
m_val  = float(input("Mass m (kg): ") or "1.0")
c_val  = float(input("Damping c (N*s/m): ") or "2.0")
k_val  = float(input("Spring k (N/m): ") or "20.0")
F0_val = float(input("External force F0 (N): ") or "10.0")
x0_val = float(input("Initial displacement x0 (m): ") or "0.0")
v0_val = float(input("Initial velocity v0 (m/s): ") or "0.0")

# Solve and plot
solve_mass_spring_damper(m_val, c_val, k_val, F0_val, x0_val, v0_val)
```
